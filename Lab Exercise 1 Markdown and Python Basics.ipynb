# Python Basics

Learn the basics of Python and its data types

## Input and Output

```python
# Input name and display name
name = input("Enter name: ")
print("Your name is ", name) # first version of the print
print("\n" + name +  " is your name") # second version of the print
print (f"{name} is the name, soon it will be a legend!") #third version of the print

age = 22 #Integer data type
print(f"Your age is {age}") #you cannot combine integer to string
print(f"\nYour age is\t {age}") # using \n and \t
```

## Data Types

### String

```python
name  = "Dan Lloyd Cabanilla"
print(f"Your name is {name}")

# check data type
print(type(name))
```

### Integer

```python
number = 5
print(f"Number is {number}")

# check data type
print(type(number))
```

### Float

```python
decimal_number = 12.123456
print(f"Number with decimal number is {decimal_number}")

# check data type
print(type(decimal_number))
```

### Boolean

```python
is_logged_in = True
print(is_logged_in)
print(is_logged_in == False)

# check data type
print(type(is_logged_in))
```

### List (mutable - it can be changed)

```python
year_level = [1, 2, 3] # square bracket
print(f"Current year level list is {year_level}")

# append another year: 4
year_level.append(4)
print(f"Updated year level is {year_level}")

# check data type
print(type(year_level))
```

### Tuple (immutable - it cannot be changed)

```python
dimensions = (10,20) # uses parentheses
print(dimensions)

# check data type
print(type(dimensions))
```

### Dictionary

```python
# uses ley-value pair, curly braces
student = { name: "Dan", age: 22, "program": "BSCS"} # uses curly braces
print(f'Student name is {student[name]}')
print(f'Age is {student[age]}')
print(f'Program is {student["program"]}')

# check data type
print(f"Student is a {type(student)}")
```

### Set

```python
setNumbers = {1, 2, 3, 3, 4} # uses curly braces;
print(setNumbers)

# check data type
print(type(setNumbers))
```
